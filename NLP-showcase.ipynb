### Inference script

To make an inference call src/inference as a module, this has several options of operation controlled by flags when calling the script. Use the '--help' flag to see them all.

```python
!python3 -m src.inference --help
```

The "Hello world" mode is the --demo flag, where the ensemble will predict the sentiment of a fixed set of sentences.

The --data_path flag is used to specify the path that contains all the pretrained weights. If mounting a drive, specifying the absolute path is recommended.

```python
!python3 -m src.inference --demo --data_path /content/drive/MyDrive/Carrera/Decimo/nlp/sentiment/data
```

You can also supply your own sentences by using the "--input" flag, and then passing a string of all the sentences separated by a "&&" (logical and) character.

```python
!python3 -m src.inference --input "Today is an amazing day.&&I didn't have a great time" \
                      --data_path /content/drive/MyDrive/Carrera/Decimo/nlp/sentiment/data
```

Notice how the final result marks both as positive, likely because the second sentence had the word "great" in it, however, if you look at the breakdown of results, bert correctly predicts the negative connotation.

### Voice recognition

The "--voice" flag allows the user to record his/her own sentence if the argument 'record' is provided, however, google colab doesn't allow for this functionality because the microphone is not accesible. A jupyter notebook will allow this behaviour.

Another option is to perform transcription over a pre-recorded audio WAV file. To do this, pass the path to the file as an argument. The repo comes with two examples: data/examples/positive.wav and data/examples/negative.wav.

```python
!python3 -m src.inference --voice data/examples/positive.wav \
                          --data_path /content/drive/MyDrive/Carrera/Decimo/nlp/sentiment/data
```

```python
!python3 -m src.inference --voice data/examples/negative.wav \
                          --data_path /content/drive/MyDrive/Carrera/Decimo/nlp/sentiment/data
```

Again, notice how the ensemble came to the conclusion that the sentence has a positive classification, but our strongest model (bert) correctly predicted negative.
